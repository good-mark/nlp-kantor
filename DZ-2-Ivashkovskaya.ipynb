```python
%pylab inline
import pandas as pd
import numpy as np
import sklearn
from matplotlib import pyplot as plt
import nltk
from nltk.collocations import *
import re
```

```python
dataset = pd.read_csv("Emails.csv", delimiter=",")
dataset.head()
```

```python
#Предобработка
texts = list(dataset["RawText"])
print texts[11]

stopwords = ['the', 'for', 'from', 'with', 'and', 'or', 'has', 'have', 'been', 'are', 'they', 'you', 'unclassified',
             'will', 'would', 'not', 'did', 'more', 'than', 'that', 'said', 'just', 'today',
            'january', 'february', 'march', 'april', 'may', 'june', 'july', 'august', 
             'september', 'october', 'november', 'december', 'dept', 
             'sunday', 'monday', 'tuesday', 'wednesday', 'thursday', 'friday', 'saturday']


for i, text in enumerate(texts):
    text = text.replace('Original Message', '').replace('U.S. Department of State', '').replace('Doc No.', '')
    text = text.replace('RELEASE IN', '').replace('Case No.', '')
    text = text.replace('STATE DEPT. - PRODUCED TO HOUSE SELECT BENGHAZI COMM.', '')
    text = text.replace('SUBJECT TO AGREEMENT ON SENSITIVE INFORMATION & REDACTIONS. NO FOIA WAIVER.', '')
    words = re.split("[\s\.,-]", text)
    #if i == 0:
        #print words
    prepared_words = [word.lower() for word in words if len(word) > 2 and word.isalpha() and word.lower() not in stopwords]
    if i == 0:
        print prepared_words
    #text = re.sub(r'[^\w]', ' ', text)
    texts[i] = prepared_words
```

Делим исходный текст на токены по пробелам и знакам препинания. Отбрасываем короткие токены, некоторые союзы-предлоги и токены, в которых встречаются цифры или какие-то небуквенные символы, чтобы отсечь часть метаинформации. Приводим оставшиеся слова к нижнему регистру.

```python
#Частотные биграммы
from sklearn.feature_extraction.text import CountVectorizer


vectorizer = CountVectorizer(ngram_range=(2,2))
X = vectorizer.fit_transform([u' '.join(text) for text in texts])

X = X.sum(axis=0)

terms = {} 
for z in vectorizer.vocabulary_:
    terms[vectorizer.vocabulary_[z]] = z

for k in xrange(20): 
    j = np.unravel_index(X.argmax(), X.shape)[1]
    print terms[j], X[0, j]
    X[0, j] = 0
```

```python
#Коллокации из nltk
bigram_measures = nltk.collocations.BigramAssocMeasures()
finder = BigramCollocationFinder.from_documents(texts)
res = finder.score_ngrams(bigram_measures.raw_freq)
print res[:10]
res = finder.score_ngrams(bigram_measures.pmi)
print res[:10]
```

```python
#Кластеризация
from sklearn.cluster import KMeans
from sklearn.feature_extraction.text import TfidfVectorizer
connected_texts = [" ".join(text) for text in texts]
counter = TfidfVectorizer(strip_accents='unicode', ngram_range=(2,2), stop_words='english', max_df=0.99, min_df=0.001)
A = counter.fit_transform(connected_texts).toarray()

from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import Normalizer
from sklearn.decomposition import TruncatedSVD

print A.shape, len(texts)
svd = TruncatedSVD(50)
normalizer = Normalizer(copy=False)
lsa = make_pipeline(svd, normalizer)
A = lsa.fit_transform(A)

km = KMeans(n_clusters=50, init='k-means++', max_iter=500, n_init=100)
result = km.fit_predict(A)
```
