```python
# Пункты 1-3
labels = []
texts = []
with open("smsspamcollection/SMSSpamCollection.txt") as f:
    for line in f:
        values = line.split("\t")
        if values[0] == "spam":
            labels.append(1)
        else:
            labels.append(0)
        texts.append(values[1].strip())
```

```python
for i in range(len(labels)):
    print labels[i], '!!!', texts[i]
```

```python
# Пункт 4
import numpy as np
import sklearn
from sklearn.feature_extraction.text import CountVectorizer

feature_extractor = sklearn.feature_extraction.text.CountVectorizer()
features = feature_extractor.fit_transform(texts)
```

```python
for row in features:
    print row
```

```python
# Пункт 5
from sklearn.linear_model import LogisticRegression
from sklearn.cross_validation import cross_val_score

scores = cross_val_score(LogisticRegression(), features, labels, n_jobs=-1, cv=10, scoring='f1')
print np.mean(scores)
```

```python
# Пункт 6
model = LogisticRegression()
model.fit(features, labels)

examples = []
examples.append("FreeMsg: Txt: CALL to No: 86888 & claim your reward of 3 hours talk time to use from your phone now! Subscribe6GB")
examples.append("FreeMsg: Txt: claim your reward of 3 hours talk time")
examples.append("Have you visited the last lecture on physics?")
examples.append("Have you visited the last lecture on physics? Just buy this book and you will have all materials! Only 99$")
examples.append("Only 99$")

examples_features = feature_extractor.transform(examples)
results = model.predict(examples_features)
print results
```

```python
# Пункт 7
ngarms_params = [(2,2), (3,3), (1,3)]
for ngrams in ngarms_params:
    print 'params is', ngrams
    feature_extractor = sklearn.feature_extraction.text.CountVectorizer(ngram_range=ngrams)
    features = feature_extractor.fit_transform(texts)
    scores = cross_val_score(LogisticRegression(), features, labels, n_jobs=-1, cv=10, scoring='f1')
    print np.mean(scores)
```

```python
# Пункт 8
from sklearn.naive_bayes import MultinomialNB

for ngrams in ngarms_params:
    feature_extractor = sklearn.feature_extraction.text.CountVectorizer(ngram_range=ngrams)
    features = feature_extractor.fit_transform(texts)
    scores = cross_val_score(MultinomialNB(), features, labels, n_jobs=-1, cv=10, scoring='f1')
    print np.mean(scores)
```

```python
# Пункт 9
from sklearn.feature_extraction.text import TfidfVectorizer

feature_extractor = sklearn.feature_extraction.text.TfidfVectorizer()
features = feature_extractor.fit_transform(texts)
scores = cross_val_score(LogisticRegression(), features, labels, n_jobs=-1, cv=10, scoring='f1')
print 'tf-idf', np.mean(scores)
```

На tf-idf качество упало.
